```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from scipy.io import loadmat
import numpy as np
from matplotlib.colors import LogNorm
```

#Load *MNISTmini* dataset

```python
# Load the MNISTmini dataset
data_path = 'data/MNISTmini.mat'
mnist_data = loadmat(data_path)

# Extract features and labels
# The dataset contains train_fea1, train_gnd1, test_fea1, test_gnd1
# We'll combine them and then split according to project requirements
train_features = mnist_data['train_fea1']  # (60000, 100)
train_labels = mnist_data['train_gnd1'].flatten()  # (60000,)
test_features = mnist_data['test_fea1']  # (10000, 100)
test_labels = mnist_data['test_gnd1'].flatten()  # (10000,)

# Combine all data (we need images 1-3000 total)
# Take first 3000 images from combined dataset
all_features = np.vstack([train_features, test_features])  # (70000, 100)
all_labels = np.hstack([train_labels, test_labels])  # (70000,)

# Split according to project requirements:
# Training: images 1-1000 (indices 0-999)
# Validation: images 1001-2000 (indices 1000-1999)
# Test: images 2001-3000 (indices 2000-2999)
X_train = all_features[0:1000]
y_train = all_labels[0:1000]

X_val = all_features[1000:2000]
y_val = all_labels[1000:2000]

X_test = all_features[2000:3000]
y_test = all_labels[2000:3000]

# Verify the splits
print(f"Training set: {X_train.shape}, labels: {y_train.shape}, unique labels: {np.unique(y_train)}")
print(f"Validation set: {X_val.shape}, labels: {y_val.shape}, unique labels: {np.unique(y_val)}")
print(f"Test set: {X_test.shape}, labels: {y_test.shape}, unique labels: {np.unique(y_test)}")
```

#Zero-mean (standardization)

```python
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

# Then train logistic regression on scaled data

```python
model = LogisticRegression(
    C=1.0,
    penalty='l2', 
    solver='liblinear', 
    max_iter=1000
)

model.fit(X_train_scaled, y_train)

#Predicting
y_pred = model.predict(X_test_scaled)

#Evaluating
```
